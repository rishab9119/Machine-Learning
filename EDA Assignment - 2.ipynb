# Q3. What is the range of prices in the dataset? What is the minimum and maximum price?

```python
df['Price'].describe()
```

Formula to calculate range of X in the dataset:

range(X) = Max(X) - Min(X)

```python
range_of_prices = df['Price'].max() - df['Price'].min()
print('The range of prices in the dataset is {}'.format(range_of_prices))
```

```python
print('The minimum price is {} and the maximum price is {}'.format( df['Price'].min(), df['Price'].max()) )
```

After calculating the range of a dataset, we can use it to gain insights into the spread of the data.

If the range is small, it suggests that the values in the dataset are tightly clustered around the mean or median, indicating low variability in the data. On the other hand, if the range is large, it suggests that the values in the dataset are more spread out, indicating high variability in the data.

In the case of flight prices, a small range may suggest that there is little variation in the prices of flights, whereas a large range may suggest that the prices of flights vary widely. Additionally, examining the range in combination with other measures of central tendency and dispersion, such as the mean, median, and standard deviation, can provide a more complete picture of the distribution of flight prices in the dataset.
